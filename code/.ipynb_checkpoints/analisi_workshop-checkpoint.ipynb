```python
#remove 0 shap values and sort 
points = dict()
for r in shaps:
    sorted(shaps[r], key = lambda x : x[0] )
    shaps[r].sort(reverse=True)
    points[r] = list()
    for i in range(0, len(shaps[r])):
        #points[r].append((i, risultati[r][i]))
        points[r].append((i, shaps[r][i][0], shaps[r][i][1]))
```

```python
import math
#unire freq e shap values:la frequenza di location con shap values alto e anche alta?
risultato = dict()
for i in range(0, len(points.keys())):
    flag = False
    risultato[i] = list()
    f = freq.loc[freq['uid'] == users[i]]
    for e in range(0,len(points[i])):
        p = eval(points[i][e][2])
        count = 0
        if p != 0 and count<6:
            if math.isclose(f['lat_home'], p[0]) & math.isclose(f['lng_home'], p[1]):
                flag = True
                risultato[i].append('home')
                risultato[i].append(f['entropy_home'].iloc[0])
                risultato[i].append(f['raw_home_freq'].iloc[0])
            elif math.isclose(f['lat_work'], p[0]) & math.isclose(f['lng_work'], p[1]):
                flag = True
                risultato[i].append('work')
                risultato[i].append(f['entropy_work'].iloc[0])
                risultato[i].append(f['raw_work_freq'].iloc[0]) 
            elif math.isclose(f['lat_least'], p[0]) & math.isclose(f['lng_least'], p[1]):
                flag = True
                risultato[i].append('least')
                risultato[i].append(f['entropy_least'].iloc[0])
                risultato[i].append(f['raw_least_freq'].iloc[0]) 
            else:    
                for k in range(2, 20):
                    if math.isclose(f['lat_'+str(k)], p[0]) & math.isclose(f['lng_'+str(k)], p[1]):
                        flag = True
                        risultato[i].append(str(k))
                        risultato[i].append(f['entropy_'+str(k)].iloc[0])
                        risultato[i].append(f[str(k)].iloc[0])  
            count += 1
```
